```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split,GridSearchCV
import hmmlearn
from hmmlearn import hmm
```

# **Data**

```python
x = pd.read_csv ('/content/drive/MyDrive/Colab Notebooks/Data Mining/ETS_HiddenMarkov/df_oversampled.csv')
x
```

```python
X=x.drop(["id","Swarm_Behaviour"],axis=1)
Y=x["Swarm_Behaviour"]
```

## **Split Data**

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=101)
```

# **Hidden Markov Model (Training Data)**

```python
# Build the HMM model
model = hmm.GaussianHMM(n_components = 3, covariance_type = "diag", n_iter = 50, random_state = 42)
ytrain = model.fit(X_train)
```

```python
# Predict the hidden states corresponding to observed X.
ypred = model.predict(X_test)
states = pd.unique(ypred)
```

# **Testing Data**

```python
print(classification_report(Y_test, ypred))
```
